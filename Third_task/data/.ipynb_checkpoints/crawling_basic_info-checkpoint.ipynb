```python
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time
```

# 21대 국회의원선거 후보자 데이터 크롤링

```python
all_data = []
columns_list = ['시도',
'선거구명',
 '사진',
 '기호',
 '소속정당명',
 '성명(한자)',
 '성별',
 '생년월일(연령)',
 '주소',
 '직업',
 '학력',
 '경력',
 '재산신고액(천원)',
 '병역신고사항(본인)',
 '납세실적(천원)-납부액',
 '납세실적(천원)-최근5년간체납액',
 '납세실적(천원)-현체납액',
 '전과기록유무(건수)',
 '입후보횟수',
'공약링크']
```

```python
driver = webdriver.Chrome('chromedriver')
```

```python
url = 'http://info.nec.go.kr/main/showDocument.xhtml?electionId=0020200415&topMenuId=CP&secondMenuId=CPRI03'

driver.get(url)
time.sleep(2)

driver.find_element_by_xpath('//*[@id="electionId2"]').click()
time.sleep(2)

for i in range(2,19):
    driver.find_element_by_xpath('//*[@id="cityCode"]/option['+str(i)+']').click()
    time.sleep(2)
    check = 2
    while check>=2:
        try:
            driver.find_element_by_xpath('//*[@id="sggCityCode"]/option['+str(check)+']').click()
            time.sleep(2)
            driver.find_element_by_xpath('//*[@id="spanSubmit"]/input').click()
            time.sleep(2)
            html=driver.page_source
            soup=BeautifulSoup(html,'html.parser')

            crawling_electionName = soup.select_one('#electionName').text[1:-1]
            crawling_city = soup.select_one('#cityName').text[1:-1]
            crawling_sgg_city = soup.select_one('#sggCityName').text[1:-1]
            
            div = soup.select_one("#table01")
            data = pd.read_html(str(div))[0].values
             
            for i in range(len(data)):
                source_html = soup.select_one('#table01 > tbody > tr:nth-child('+str(i+1)+') > td:nth-child(5)').find_all('a', href=True)[1]['href'][28:-2].split(',')
                promise_html = "http://policy.nec.go.kr/plc/popup/initUMAPopup.do?sgId="+source_html[0].split('.')[0][1:-1].strip("0")+"&subSgId="+source_html[1].split('.')[0].split('+')[0][1:-1]+source_html[1].split('.')[0].split('+')[1][1:-1].strip("0")+"&huboid="+source_html[2][1:-1]
                append_data = np.insert(data[i],0, crawling_city).tolist()
                append_data = np.append(append_data, promise_html)
                all_data.append(append_data)   
                
            check += 1
            print("현재 가져온 데이터는 "+ crawling_electionName, crawling_city, crawling_sgg_city + " 이며 총 "+str(len(data))+"명 입니다.")
            
            time.sleep(2)
            
        except NoSuchElementException:
            check = 1
            time.sleep(2)
            
driver.quit()
```

```python
all_df = pd.DataFrame(data=all_data, columns=columns_list)
```

```python
all_df.shape
```

# 데이터 전처리

```python
all_df['성명(한자)'] = all_df['성명(한자)'].map(lambda x : x[:-5])
all_df['나이'] = all_df['생년월일(연령)'].map(lambda x : x[-4:-1])
all_df['생년원일'] = all_df['생년월일(연령)'].map(lambda x : x[:-5])
```

```python
all_df.to_csv('21대_국회의원선거_후보자_데이터.csv', index=False)
```
